```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
```

```python
axis = pd.read_csv("AXISBANK.csv")
axis.info()
axis
```

```python
plt.figure(figsize=(10,5))
plt.scatter(axis['Volume'], axis['Close'])
plt.xlabel('Volume')
plt.ylabel('closing price')

plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(axis['Close'])
plt.title('Axis bank stock closing price over time')
plt.xlabel('Date')
plt.ylabel('closing price')

plt.figure(figsize=(5,3))
plt.plot(axis['Volume'], color = 'orange')
plt.title('Axis bank stock volume traded over time')
plt.xlabel('Date')
plt.ylabel('Volume traded')
plt.show()
```
